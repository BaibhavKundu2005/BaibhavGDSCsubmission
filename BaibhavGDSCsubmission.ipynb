```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('/Users/bijit/Downloads/train (7).csv')
test = pd.read_csv('/Users/bijit/Downloads/test (4).csv')
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
X_train = train.drop(columns=['Unnamed: 0','target'])
y_train = train['target']
```

```python
X_test = test.drop(columns = 'Unnamed: 0')
```

```python
model = LinearRegression()
```

```python
rmse_scorer = make_scorer(mean_squared_error, squared=False)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_train, y_train, cv=kf, scoring=rmse_scorer)
```

```python
cv_scores
```

```python
np.mean(cv_scores)
```

```python
model.fit(X_train, y_train)
```

```python
test_pred = model.predict(X_test)
```

```python
submission = pd.read_csv("/Users/bijit/Downloads/sample.csv")
submission['target'] = test_pred
submission.to_csv("submission.csv",index = False)
```
